```python
# Define class names (same as your dataset folders)
classes = [
    "apple_pie",
    "cheesecake",
    "chicken_curry",
    "french_fries",
    "fried_rice",
    "hamburger",
    "hot_dog",
    "ice_cream",
    "omelette",
    "pizza",
    "sushi"
]

# Define calorie mapping (approx values per serving)
calorie_map = {
    "apple_pie": 320,
    "cheesecake": 350,
    "chicken_curry": 250,
    "french_fries": 312,
    "fried_rice": 330,
    "hamburger": 295,
    "hot_dog": 290,
    "ice_cream": 207,
    "omelette": 154,
    "pizza": 285,
    "sushi": 200
}

print("Classes loaded:")
print(classes)

print("\nCalorie map loaded:")
for k, v in calorie_map.items():
    print(f"{k} -> {v} kcal")
```

```python
path = Path("/content/calorie_dataset/food11")
train_path = path / "train"
test_path = path / "test"

print(path)
print(train_path.exists(), test_path.exists())
print("Train image count:", len(get_image_files(train_path)))
print("Test image count :", len(get_image_files(test_path)))
```

```python
dls = ImageDataLoaders.from_folder(
    path,
    train="train",
    valid="test",
    item_tfms=Resize(224),
    bs=32
)
```

```python
learn = vision_learner(dls, resnet50, metrics=accuracy)
```

```python
learn.fit_one_cycle(8, 1e-3)
```

```python
learn.save('food11-model')
```

```python
learn.load('food11-model')
```

```python
interp = Interpretation.from_learner(learn)
```
